```python
import numpy as np
```

```python
import spectra as s
```

```python
import matplotlib.pyplot as plt
```

```python
uppera = np.array([])
upperb = np.array([])
```

```python
for i in range(10):
    npz = np.load('_upper/_upper' + str(i) + '.npz')
    data = npz['data']
    data.shape = (2, -1)
    uppera = np.append(uppera, data[0]*0.1/(2**15))
    upperb = np.append(upperb, 1j*(data[1]*0.1/(2**15)))
```

```python
uppera
```

```python
datac = uppera + upperb
datac.shape = (-1, 16000)
ft_data = np.fft.fft(datac, axis=1)
meanspec = np.mean(np.abs(ft_data)**2, axis=0)
```

```python
print(meanspec)
```

```python
plt.plot(meanspec)
```

```python
%matplotlib notebook
s.plot_power(meanspec, 'meanupper')
```

```python
np.savez('meanspecupper', data=meanspec)
```

```python
lowera = np.array([])
lowerb = np.array([])
```

```python
for i in range(10):
    npz = np.load('retakelower/retakelower' + str(i) + '.npz')
    data = npz['data']
    data.shape = (2, -1)
    lowera = np.append(lowera, data[0]*100/(2**15))
    lowerb = np.append(lowerb, 1j*(data[1]*100/(2**15)))
```

```python
datacl = lowera + lowerb
datacl.shape = (-1, 16000)
ft_datal = np.fft.fft(datacl, axis=1)
meanspecl = np.mean(np.abs(ft_datal)**2, axis=0)
```

```python
s.plot_power(meanspecl, 'meanlower')
```

```python
np.savez('meanspeclower', data=meanspecl)
```

```python
half = len(meanspecl)/2
onl = meanspecl[:int(half)]
onu = meanspec[int(half):]
son = np.append(onl, onu)
```

```python
s.plot_power(son, 'son')
```

```python
half = len(meanspecl)/2
offu = meanspec[:int(half)]
offl = meanspecl[int(half):]
soff = np.append(offl, offu)
```

```python
s.plot_power(soff, 'soff')
```

```python
s.plot_power(son/soff, 'line shape')
```
